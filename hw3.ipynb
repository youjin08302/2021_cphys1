## 1. Leibniz Pi formula

1. 식을 구현하는 함수 구현

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(n):
    f=0
    for i in range(1, n+1):
        f=f+((-1)**(i+1))*1/(2*i-1)
    return f
```

2. 이 식을 이용하여 그래프 그리기

```python
def f(n):
    f=0
    for i in range(1, n+1):
        f=f+((-1)**(i+1))*1/(2*i-1)
    return f
```

```python
a = [f(n) for n in range (1,101)]
plt.plot(a,color='skyblue',label='Leibniz Pi')
plt.hlines(np.pi/4,0,100,color='orange',label='π/4')
plt.legend()
```
